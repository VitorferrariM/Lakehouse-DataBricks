```python
import streamlit as st
import duckdb
import pandas as pd
%pip install streamlit
%pip install duckdb
# LÃª o Parquet
df = pd.read_parquet("nasa_gold.parquet")

# Conecta ao DuckDB
con = duckdb.connect("nasa.duckdb")
con.register("nasa_weather_df", df)
con.execute("CREATE TABLE IF NOT EXISTS nasa_weather AS SELECT * FROM nasa_weather_df")

# Consulta os dados
df = con.execute("SELECT * FROM nasa_weather").fetchdf()

# Interface Streamlit
st.title("ğŸŒ Painel de Temperaturas da NASA")

# Filtro por estaÃ§Ã£o
season = st.selectbox("Escolha a estaÃ§Ã£o:", df["season"].unique())
filtered = df[df["season"] == season]

# GrÃ¡fico de temperaturas
st.subheader(f"ğŸ“ˆ Temperaturas para a estaÃ§Ã£o: {season}")
st.line_chart(filtered[["avg_temp_c", "min_temp_c", "max_temp_c"]])

# GrÃ¡fico de desvio padrÃ£o
st.subheader("ğŸ“Š Desvio padrÃ£o da temperatura")
st.bar_chart(filtered["stddev_temp_c"])

# GrÃ¡fico de observaÃ§Ãµes
st.subheader("ğŸ“‹ NÃºmero de observaÃ§Ãµes")
st.bar_chart(filtered["observations"])

# Tabela completa
st.subheader("ğŸ” Dados filtrados")
st.dataframe(filtered)
```
