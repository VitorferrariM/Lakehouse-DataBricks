```python
import duckdb
 import os

# Conectar ao DuckDB
con = duckdb.connect()

# Instalar e carregar extens√£o httpfs
con.execute("INSTALL httpfs;")
con.execute("LOAD httpfs;")

# Configurar credenciais da AWS
con.execute("SET s3_region='us-east-1';")
con.execute("SET s3_access_key_id='AWS_ACCESS_KEY_ID';")
con.execute("SET s3_secret_access_key='AWS_SECRET_ACCESS_KEY';")


# Caminho HTTP direto para o arquivo Parquet
http_url = "https://nasa-weather-bronze.s3.amazonaws.com/bronze/nasa_weather/nasa_insight_2025-11-02.parquet"

con.execute("""
CREATE TABLE bronze_data AS
SELECT * FROM read_parquet('s3://nasa-weather-bronze/bronze/nasa_weather/nasa_insight_2025-11-02.parquet');
""")

# Ver estrutura da tabela
print(con.execute("DESCRIBE bronze_data").fetchdf())

# Ver os primeiros registros
print(con.execute("SELECT * FROM bronze_data LIMIT 10").fetchdf())


```
